```python
import pandas as pd
```

```python
df = pd.read_csv('../../dataset/data.csv')
df.head()
```

```python
pearson = df['ammonia'].corr(df['chlorophyll'], method='pearson')
spearman = df['ammonia'].corr(df['chlorophyll'], method='spearman')
kendall = df['ammonia'].corr(df['chlorophyll'], method='kendall')

(pearson, spearman, kendall)
```

```python
from matplotlib import pyplot as plt

plt.plot(df['chlorophyll'][:50], label='chlrophyll')
plt.plot(df['ammonia'][:50], label='ammonia')

plt.legend()
plt.show()
```
